```python
"""
Question 1:
Write a program which takes 2 digits, X,Y as input and generates a 2-dimensional array.
The element value in the i-th row and j-th column of the array should be i*j.
Note: i=0,1.., X-1; j=0,1,¡­Y-1.
Expected Input:
3,5
Expected Output:
[[0, 0, 0, 0, 0], [0, 1, 2, 3, 4], [0, 2, 4, 6, 8]] 
Note:
In case of taking data from the user, it should be in a comma-separated form.
"""

import numpy as np

i, j = map(int, input("Enter i, j: ").split(','))

array = np.zeros((i, j))

for a in range(i):
  for b in range(j):
    array[a][b] = a * b
  
array = array.astype(int)

print(array.tolist())
```

```python
""" 
Question 7:
Write a program that computes the net amount of a bank account based a transaction log from console input. The transaction log format is shown as following:
D 100
W 200
D means deposit while W means withdrawal.
Expected Input:
D 300
D 300
W 200
D 100
Expected Output:
500
Note:
In case of taking data from the user, it should be in a comma-separated form.
"""

""" Note: I am not splitting on the newline character because of how python is taking 
input in the jupyter notebook"""

account = input("Enter transactions")

account = account.split(' ')

net = 0

for i in range(0, len(account), 2):
  if account[i] == 'D':
    net += int(account[i+1])
  elif account[i] == 'W':
    net -= int(account[i+1])

print(net)
```

```python
""" 
Question 9:
You are required to write a program to sort the (name, age, height) tuples by ascending order where name is string, age and height are numbers. The tuples are input by console. The sort criteria is:
1: Sort based on name;
2: Then sort based on age;
3: Then sort by score.
The priority is that name > age > score.
Expected Input:
Tom,19,80
John,20,90
Jony,17,91
Jony,17,93
Json,21,85
Expected Output:
[('John', '20', '90'), ('Jony', '17', '91'), ('Jony', '17', '93'), ('Json', '21', '85'), ('Tom', '19', '80')]
Note:
In case of taking data from the user, it should be in a comma-separated form.
"""

from operator import itemgetter

data = []

line = input("Enter a tuple (name, age, height)")

line = line.split(' ')

for l in line:
  name, age, height = l.split(',')
  
  data.append((name, int(age), int(height)))

# Sort the data
# itemgetter(0, 1, 2) means sort by name, then age, then height
data.sort(key=itemgetter(0, 1, 2))

# Print the sorted data
print(data)
```

```python
""" 
Question 10:
A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The numbers after the direction are steps.  
The trace of robot movement is shown as the following:
Expected Input:
UP 5
DOWN 3
LEFT 3
RIGHT 2
Expected Output:
Compute the distance from current position after a sequence of movement and original point. If the distance is a float, then just print the nearest integer.
Note:
In case of taking data from the user, it should be in a comma-separated form.
"""

movements = input("Enter movements")

movements = movements.split(' ')

x = 0
y = 0

for dir in movements:
  if dir == 'UP':
    y += int(movements[movements.index(dir) + 1])
  elif dir == 'DOWN':
    y -= int(movements[movements.index(dir) + 1])
  elif dir == 'LEFT':
    x -= int(movements[movements.index(dir) + 1])
  elif dir == 'RIGHT':
    x += int(movements[movements.index(dir) + 1])

# since starting position is 0,0, we don't need to subtract x_0, y_0 from x, y
print(round(np.sqrt(x**2 + y**2)))
```

```python
""" 
Question 11:
Find the continuous occurrence of the string.
Expected Input:
Aabbcdeefffaabbcc
Expected Output:
a2b2c1d1e2f3a2b2c2
"""

string = input("Enter a string").lower()

# go over eacht erm in the string
# counter = number of terms equal to i-1 in sequence
# if term is equal to term before, increment counter
# if encountering a new term, add the counter to the end and add the new character tot he list

output_string = ""

counter = 1

prev = string[0]
output_string += prev

for i in range(1, len(string)):
  if string[i] == prev:
    counter += 1
  else:
    output_string += str(counter) + string[i]
    prev = string[i]
    counter = 1

output_string += str(counter)

print(output_string)
```

```python
""" 
Question 14:
Find the minimum possible denominations for given valid currency.
(No of currencies used should be minimum)
Expected Input 1:
valid_currency: [1,2,5,10,20,50,100,200,500,2000]
Money: 210
Expected Output:
200-1
10-1
Expected Input 2:
valid_currency: [1,2,5,10,20,50,100,200,500]
Money: 556
Expected Output:
500-1
50-1
5-1
1-1
Expected Input 3:
valid_currency: [1,2,5,10,20,50,100,200,500,2000]
Money: 2000
Expected Output:
2000-1
Expected Input 4:
valid_currency: [1,2,5,10,20,50,100,500,1000]
Money: 210
Expected Output:
100-2
10-1
Expected Input 5:
valid_currency: [1,2,5,10,20,50,100,200,500,1000]
Money: 2000
Expected Output:
1000-2
"""

""" 

I am going to write this as answer like the complete change making problem, as opposed to the greedy algorithm that can solve these cases optimally

"""
import numpy as np

inputs = input("Enter valid currencies: ").split(',')
currencies = [int(c) for c in inputs]
money = int(input("Enter money:"))

table = [0] + [np.inf] * money

coins_used = [0] * (money + 1)

for c in currencies:
    for i in range(c, money + 1):
        if table[i] > table[i - c] + 1:
            table[i] = table[i - c] + 1
            coins_used[i] = c

if table[money] == np.inf:
    print("It's not possible to make change for the given amount with the provided currencies.")
else:
    print("The coins used are:")
    change = money
    dictionary = {}
    for i in range(len(currencies)):
        dictionary[currencies[i]] = 0

    while change > 0:
        dictionary[coins_used[change]] += 1
        change -= coins_used[change]
    
    keys = sorted(dictionary.keys(), reverse=True)
    
    for key, value in dictionary.items():
        if value > 0:
            print(f"{key}-{value}")
```

```python
""" 
Question 15:
There is a bus travelling from Town A to Town B. There are n stops between them and bus has to make m stops.
Find the numbery of ways in the travel so that no stop is consecutive
Expected Input 1:
n=12
m=4
Expected Output:
Output :126
Expected Input 2:
n = 16
s = 5
Expected Output:
792
"""

n = int(input("Enter the number of stops between the towns:"))
m = int(input("Enter the number of stops the bus has to make:"))


dp = [[0 for _ in range(m+1)] for _ in range(n+2)]

for i in range(n+2):
    dp[i][0] = 1

for i in range(2, n+2):
    for j in range(1, min(i, m)+1):
        dp[i][j] = dp[i-1][j]
        if i > j:
            dp[i][j] += dp[i-2][j-1]

print(dp[n+1][m])
```

```python
""" 
Question 20:
In a pathology lab test, there are n number of samples for testing the health condition of a patient,
each slide has 5 components, Sugar level, Blood pressure, Heartbeat rate, weight and fat percentage,
based on input as provided by the patient's blood report.

1. Create a sample input for a healthy patient as follows:

"Sugar level":15, "Blood pressure":32, "Heartbeat rate":71, "weight":65, "fat percentage":10.

2. Get values from the user and compare inputs with healthy patient data. If the patient data is not matching with the healthy
patient’s data, provide a warning.

3. Provide difference in readings to the patient.
Expected Output:
Sugar level:56

Blood pressure:120

Heartbeat rate:45

weight:67

fat percentage:67

{'Sugar level': -41, 'Blood pressure': -88, 'Heartbeat rate': 26, 'weight': -2, 'fat percentage': -57}

Sugar level -41

The sugar level is 41 less than the ideal value

Blood pressure -88

Blood pressure is 88 less than the ideal value

Heartbeat rate 26

Heartbeat rate is 26 more than the ideal value

weight -2

weight is 2 less than the ideal value

fat percentage 57

fat percentage is 57 less than the ideal value
"""

healthy_patient = {
    "Sugar level": 15,
    "Blood pressure": 32,
    "Heartbeat rate": 71,
    "weight": 65,
    "fat percentage": 10
}

patient_data = {}
for key in healthy_patient.keys():
    patient_data[key] = int(input(f"Enter {key}: "))

differences = {}
for key in healthy_patient.keys():
    differences[key] = healthy_patient[key] - patient_data[key]
    if differences[key] != 0:
        print(f"\n{key} {differences[key]}")
        if differences[key] > 0:
            print(f"The {key} is {differences[key]} less than the ideal value")
        else:
            print(f"The {key} is {-differences[key]} more than the ideal value")

print("\nDifferences:", differences)
```

```python
""" 
Question 21:
Check whether the given number is Armstrong number or not
Armstrong number: 153 => 1^3+5^3+3^3=153 (If summing each digit to the power of number of digits results to the same number then it is a Armstrong number)

1634 => 1^4+6^4+3^4+4^4 =1634         1^4 => digit^(number of digits)
Expected Input:
1634
Expected Output:
 Armstrong number
"""

number = input("Enter a number: ")

digits = len(number)

sum = 0

for i in number:
  sum += int(i)**digits

if sum == int(number):  
  print("Armstrong number")
else:
  print("Not an Armstrong number")
```

```python
""" 
Question 23:

Find whether the given number is perfect number or not.
Any number can be perfect number in Python, if the sum of its positive divisors excluding the number itself is equal to that number.

For example, 6 is a perfect number in Python because 6 is divisible by 1, 2, 3 and 6. 
So, the sum of these values are: 1+2+3 = 6 (Remember, we have to exclude the number itself. 
That’s why we haven’t added 6 here). Some of the perfect numbers are 6, 28, 496, 8128 and 33550336 so on.
Expected Input:
28
Expected Output:
Perfect number
"""

number = int(input("Enter a number: "))

sum = 0

for i in range(number//2+1):
  if i == 0:
    continue
  if number % i == 0:
    sum += i

if sum == number:
  print("Perfect number")
else:
  print("Not a perfect number")
```
